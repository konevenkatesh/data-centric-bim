### Load Ifc file

```python
bimonto = get_ontology("bimonto.owl").load()
iprok = get_ontology("iprok.owl").load()
```

## Merge Ontologies

```python
iprok.imported_ontologies.append(bimonto)
```

Product == BuildingElement ; 
Task == Process

### Merge Ontologies with Equivalent functions

```python
task = iprok.Task 
product = iprok.Product
buildingelement = bimonto.BuildingElement
process = bimonto.Process
```

```python
task.equivalent_to.append(process)
product.equivalent_to.append(buildingelement)
```

```python
name = iprok.Name
globalid = iprok.GlobalId
name1 = bimonto.Name
globalid1 = bimonto.GlobalId
description = iprok.Description
description1 = bimonto.Description
```

```python
name.equivalent_to.append(name1)
globalid.equivalent_to.append(globalid1)
```
