### Import core libraries

**Algorithm 1:** IFC-to-BIMOnto Transformation

**Input:** `ifc_file`, `ontology_schema`
**Output:** Populated `ontology_instance`

1.  `ifc_classes` ← [`IfcWall`, `IfcSlab`, `IfcColumn`, ...]
2.  **for each** `ifc_class` **in** `ifc_classes` **do**
3.  ```
     `objects` ← Get all instances of `ifc_class` from `ifc_file`
    ```
4.  ```
     **for each** `obj` **in** `objects` **do**
    ```
5.  ```
         `class_name` ← Get simple name from `ifc_class` (e.g., "Wall")
    ```
6.  ```
         `NewClass` ← Find or create class `class_name` in `ontology_schema`
    ```
7.  ```
         `inst` ← Create instance of `NewClass` using `obj.GlobalId`
    ```
8.  ```
         Set `inst.Name` to `obj.Name`
    ```
9.  ```
         `container` ← Get spatial container of `obj`
    ```
10. ```
        Assign `inst` to `container` in `ontology_instance`
    ```
11. ```
        `qto_sets` ← Get all quantity property sets for `obj`
    ```
12. ```
        **for each** `prop` **in** `qto_sets` **do**
    ```
13. ```
            Set `inst.prop_name` to `prop_value`
    ```
14. ```
        **end for**
    ```
15. ```
    **end for**
    ```
16. **end for**
17. **return** `ontology_instance`
