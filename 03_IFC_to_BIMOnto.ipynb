### Import core libraries

```python
import ifcopenshell
import ifcopenshell.util.element as Element
from owlready2 import *
import pprint
import uuid

def create_uuid(prefix):
    return f"{prefix}_{str(uuid.uuid4())[:5]}"

# load ifc file
ifc_file = ifcopenshell.open("Duplex_House.ifc")
# load ontology
onto = get_ontology("bimonto.owl").load()

# Define a list of properties to extract from the IFC file
properties = [
    "Length", "Width", "Height", "NetVolume", "NetSideArea",
    "GrossVolume", "Perimeter", "NetAreaGrossArea", "OuterSurfaceArea",
    "CrossSectionArea", "Span", "NumberOfRiser", "RiserHeight", "TreadLength"
]

def Mapping(onto_class, ifc_class) :
    objects = ifc_file.by_type(ifc_class)
    with onto :
        for obj in objects :
            level = Element.get_container(obj).Name if Element.get_container(obj) else ''
            PredifinedType = Element.get_predefined_type(obj) if  Element.get_predefined_type(obj) else ''
            building_storey = onto["BuildingStorey"](level)
            
            inst = onto[onto_class](obj.GlobalId)
        
            inst.GlobalId = obj.GlobalId
            if inst.owlId is None: inst.owlId = create_uuid(ifc_class)
            inst.Name = obj.Name
            inst.hasBuildingStorey = building_storey
            building_storey.hasBuildingElement.append(inst)
            inst.hasPredefinedType = onto["PredefinedType"](PredifinedType)
            inst.is_a.append(onto["PredefinedType"](PredifinedType))

            Quantitysets = Element.get_psets(obj, qtos_only=True)
            # Iterate over the properties and set them to the instance
            for prop_name in properties:
                value = next((value.get(prop_name) for value in Quantitysets.values() if prop_name in value), None)
    
                if value is not None:
                    setattr(inst, prop_name, value)
                    
            # Add the instance to the ontology
            
            
            # Save the ontology to a file
            onto.save(file="ifc_bimonto.owl", format="rdfxml")
```

```python
import ifcopenshell
import ifcopenshell.util.element as Element
from owlready2 import *
import uuid

def create_uuid(prefix):
    return f"{prefix}_{str(uuid.uuid4())[:5]}"

ifc_file = ifcopenshell.open("../Models/Duplex_House.ifc")
onto = get_ontology("../Output/bimonto.owl").load()

# Define a list of properties to extract from the IFC file
properties = [
    "Length", "Width", "Height", "NetVolume", "NetSideArea",
    "GrossVolume", "Perimeter", "NetAreaGrossArea", "OuterSurfaceArea",
    "CrossSectionArea", "Span", "NumberOfRiser", "RiserHeight", "TreadLength"
]


def Convert():
    ifc_classes = ["IfcColumn", "IfcBeam", "IfcWall", "IfcSlab", "IfcFooting", "IfcRoof", "IfcStair", "IfcRamp", "IfcWindow", "IfcDoor"  ]
    with onto:
        if not hasattr(onto, "BuildingElement"):
            raise Exception("Class 'BuildingElement' not found in ontology")

        for ifc_class in ifc_classes:
            objects = ifc_file.by_type(ifc_class)

            for obj in objects:
                class_name = ifc_class.replace("Ifc", "")  # e.g., Column

                # Try to get or create subclass of BuildingElement
                NewClass = onto.search_one(iri="*" + class_name)
                if NewClass is None or not isinstance(NewClass, ThingClass):
                    NewClass = types.new_class(class_name, (onto.BuildingElement,))
                    print(f" Created new class: {class_name}")
                # Create instance with GlobalId
                inst = NewClass(obj.GlobalId)
                inst.GlobalId = obj.GlobalId
                inst.owlId = inst.owlId or create_uuid(ifc_class)
                inst.Name = obj.Name
                # Storey assignment
                level = Element.get_container(obj).Name if Element.get_container(obj) else ''
                storey = onto["BuildingStorey"](level)
                inst.hasBuildingStorey = storey
                storey.hasBuildingElement.append(inst)
                # PredefinedType
                pred_type = Element.get_predefined_type(obj) or ""
                inst.hasPredefinedType = onto["PredefinedType"](pred_type)
                # QTO Properties
                Quantitysets = Element.get_psets(obj, qtos_only=True)
                for prop_name in properties:
                    value = next((value.get(prop_name) for value in Quantitysets.values() if prop_name in value), None)
                    if value is not None:
                        setattr(inst, prop_name, value)

Convert()

 # Save the ontology
onto.save(file="../Output/ifc_bimonto.owl", format="rdfxml")
```
